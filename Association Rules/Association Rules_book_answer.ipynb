# Apriori Algorithm

```python
freq_itemset1=apriori(df,min_support=0.1,use_colnames=True)
import warnings
warnings.filterwarnings('ignore')
freq_itemset1
```

```python
rules1=association_rules(freq_itemset1,metric='lift',min_threshold=1)
rules1
rules1.sort_values('lift',ascending=False).head(15)
```

```python
sns.scatterplot(x='support',y='confidence', data= rules1)
```

- support value = 10% , confidence / min_threshold = 1

```python
freq_itemset2=apriori(df,min_support=0.2,use_colnames=True)
freq_itemset2
```

```python
rules2=association_rules(freq_itemset2,metric='lift',min_threshold=0.7)
rules2
rules2.sort_values('lift',ascending=False).head(15)
```

```python
sns.scatterplot(x='support',y='confidence', data= rules2)
```

- support value = 20%  , confidence / min_threshold = 0.7
