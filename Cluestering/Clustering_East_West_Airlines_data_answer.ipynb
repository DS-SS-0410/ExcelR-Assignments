```python
kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(scaled_airlines_df)
```

```python
airlines_data1["Cluster"] = kmeans.predict(scaled_airlines_df)
```

```python
airlines_data1.head()
```

# Elbow Graph

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(2,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airlines_df)
    dist = kmeans.inertia_
    wcss.append(dist)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("Within a cluster sum of square")
plt.grid()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airlines_df)
    wcss.append(kmeans.inertia_)
    if i > 6:
      break
```

```python
df.head()
```

```python
airlines=df.copy()
```

```python
airlines.info()
```

```python
#Standarization (z-score)
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler()
x = stscaler.fit_transform(airlines)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=2.3, min_samples=13)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
airlines["-1_outlier"] = dbscan.labels_
```

```python
dl=dbscan.labels_
```

```python
import sklearn
sklearn.metrics.silhouette_score(x,dl)
```
